```python
import pandas as pd
import numpy as np
from pathlib import Path
import re
```

# Check for missing labels

```python
file_path = Path.cwd().joinpath("Data", "mpr_data_labelled")
```

```python
for date in pd.date_range(start="1995-05-15", end="2023-04-12"):
    try:
        df = pd.read_excel(file_path.joinpath(f"MPR-{date.strftime('%Y-%m-%d')}.xlsx"))
        missing = df['class'].isna()
        row = df[missing]
        num_missing = len(row)
        if num_missing > 0:
            print(num_missing, date.strftime('%Y-%m-%d'))
            print(f"row:{list(row.index+2)}")
        else:
            print(date.strftime('%Y-%m-%d'), "is clean")
        del df
    except:
        pass
```

# Check for miss-labelling 

* anything thats not equal to 1, 2, or 3

```python
for date in pd.date_range(start="1995-05-15", end="2023-04-12"):
    try:
        df = pd.read_excel(file_path.joinpath(f"MPR-{date.strftime('%Y-%m-%d')}.xlsx"))
        values = df['class'].apply(lambda x: None if x not in [1, 2, 3] else x)
        doubles = list(df[values.isna()].index+2)
        if len(doubles) > 0:
            print(doubles, date.strftime('%Y-%m-%d'))
        else:
            print(date.strftime('%Y-%m-%d'), "is clean")
        del df
    except:
        pass
```

```python
for date in pd.date_range(start="2019-01-01", end="2023-04-12"):
    try:
        df = pd.read_excel(file_path.joinpath(f"MPR-{date.strftime('%Y-%m-%d')}.xlsx"))
        for sentence in df.text:
            periods = sentence.count('.')
            if periods > 1:
                print(
                    list(df[df['text'] == sentence].index + 2), 
                    date.strftime('%Y-%m-%d'),
                    "\n",
                    sentence 
                )
        del df
    except:
        pass
```

```python
for date in pd.date_range(start="2017-01-01", end="2023-04-12"):
    try:
        df = pd.read_excel(file_path.joinpath(f"MPR-{date.strftime('%Y-%m-%d')}.xlsx"))
        print(list(df[df.text.isna()].index + 2), date.strftime('%Y-%m-%d'))
        del df
    except:
        pass
```
